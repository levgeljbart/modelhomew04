```python
##метод прямоугольника (https://www.articleshub.net/2023/05/metod-srednikh-pryamougolnikov-python.html)
def f(x):
    return (np.cosh(3*x))**3

def rect(f,xmin,xmax,n):
    dx=(xmax-xmin)/n
    area=0
    x=xmin
    i = 0
    while i < n:
        area+=dx*f(x)
        x+=dx
        i += 1
    return area

print('0.1', rect(f, 0, 1/3, 0.1))
print('1', rect(f, 0, 1/3, 1))
print('10', rect(f, 0, 1/3, 10))
print('100', rect(f, 0, 1/3, 100))
print('1000', rect(f, 0, 1/3, 1000))
print('10000', rect(f, 0, 1/3, 10000))
print('100000', rect(f, 0, 1/3, 100000))
print('1000000', rect(f, 0, 1/3, 1000000))
```

```python
#метод трапеций (https://www.articleshub.net/2023/05/metod-trapetsii-python_1.html)
def f(x):
    return (np.cosh(3*x))**3

def trap(xmin,xmax,n, f):
    dx=(xmax-xmin)/n
    area=0
    x=xmin
    i = 0
    while i < n:
        area+=dx*(f(x)+f(x+dx))/2
        x+=dx
        i += 1
    return area

print('0.1', trap(0, 1/3, 0.1, f))
print('1', trap(0, 1/3, 1, f))
print('10', trap(0, 1/3, 10, f))
print('100', trap(0, 1/3, 100, f))
print('1000', trap(0, 1/3, 1000, f))
print('10000', trap(0, 1/3, 10000, f))
```

```python
#метод симпсона (тоже из интернета, забыл откуда)
def f(x):
    return (np.cosh(3*x))**3

def simp(a, b, n, f):
    h = (b - a) / n
    intg = f(a) + f(b)
    i = 0
    while i < n:
        x = a + i * h
        if i % 2 == 0:
            intg += 2 * f(x)
        else:
            intg += 4 * f(x)
        i += 1
    intg *= h / 3
    return intg

print('0.1', simp(0, 1/3, 0.1, f))
print('1', simp(0, 1/3, 1, f))
print('10', simp(0, 1/3, 10, f))
print('100', simp(0, 1/3, 100, f))
print('1000', simp(0, 1/3, 1000, f))
print('10000', simp(0, 1/3, 10000, f))
print('100000', simp(0, 1/3, 100000, f))
print('1000000', simp(0, 1/3, 1000000, f))
```
